```julia
if isdir("/content/KAEM")
    cd("/content/KAEM")
elseif @__DIR__() != ""
    cd(joinpath(@__DIR__, ".."))
end

using Pkg; Pkg.activate(pwd()); Pkg.instantiate()

ENV["PERCEPTUAL"] = "false"
ENV["DEVICE"] = "gpu"

using Random, ConfParser, Lux, Reactant, Enzyme, ComponentArrays, Accessors, Statistics
using Printf, JSON

include("src/utils.jl");              using .Utils
include("src/KAEM/KAEM.jl");          using .KAEM_model, .KAEM_model.EBM_Model, .KAEM_model.GeneratorModel
include("src/KAEM/model_setup.jl");   using .ModelSetup
include("src/pipeline/optimizer.jl"); using .optimization
include("src/pipeline/data_utils.jl"); using .DataUtils: get_vision_dataset
include("src/KAEM/rng.jl");           using .HLOrng
```

```julia
function setup(; config="config/svhn_config.ini")
    rng = Random.MersenneTwister(1)
    conf = ConfParse(config); parse_conf!(conf)

    N_train = parse(Int, retrieve(conf, "TRAINING", "N_train"))
    N_test = parse(Int, retrieve(conf, "TRAINING", "N_test"))
    batch_size = parse(Int, retrieve(conf, "TRAINING", "batch_size"))
    cnn = parse(Bool, retrieve(conf, "CNN", "use_cnn_lkhood"))
    dataset_name = retrieve(conf, "TRAINING", "dataset")

    dataset, x_shape, _ = get_vision_dataset(dataset_name, N_train, N_test, batch_size; img_resize=(32,32), cnn=cnn)
    model = init_KAEM(dataset, conf, x_shape; rng=rng)
    x, _ = iterate(model.train_loader)
    x = pu(x)
    opt = create_opt(conf)
    model, opt_st, ps, sk, sl, sr = prep_model(model, x, opt; rng=rng, MLIR=false)
    sr = seed_rand(model; rng=rng)

    return model, opt_st, ps, sk, sl, sr, x
end

function bench(f, args...; warmup=3, trials=20)
    compiled = Reactant.@compile f(args...)
    for _ in 1:warmup; compiled(args...); end
    times = [(@elapsed compiled(args...)) * 1000 for _ in 1:trials]
    return compiled, times
end

"""Extract HLO op names from StableHLO IR and return sorted (op, count, %) table."""
function hlo_op_profile(hlo_str::String)
    ops = String[]
    for line in split(hlo_str, '\n')
        m = match(r"stablehlo\.(\w+)", line)
        m !== nothing && push!(ops, m.captures[1])
    end
    total = length(ops)
    counts = sort(collect(countmap(ops)); by=last, rev=true)
    return [(op, n, 100.0 * n / total) for (op, n) in counts]
end

function countmap(v)
    d = Dict{String,Int}()
    for x in v; d[x] = get(d, x, 0) + 1; end
    return d
end

function print_hlo_profile(name, profile)
    @printf("\n=== %s (%d total ops) ===\n", name, sum(x[2] for x in profile))
    @printf("%-25s %8s %8s\n", "Op", "Count", "%")
    println("-"^43)
    for (op, n, pct) in profile
        @printf("%-25s %8d %7.1f%%\n", op, n, pct)
    end
end
```

## Profile

```julia
model, opt_st, ps, sk, sl, sr, x = setup()
ebm, gen = model.prior, model.lkhood
z = pu(randn(Float32, ebm.q_size, ebm.p_size, model.batch_size))
z_samp = pu(Float32.(Array(first(model.sample_prior(model, ps, sk, sl, sr)))))

items = [
    ("ebm_forward",   (z,p,s,l) -> ebm(p,s,l,z),
        (z, ps.ebm, sk.ebm, sl.ebm)),
    ("gen_forward",   (p,s,l,z) -> gen.output_activation(first(gen.generator(p,s,l,z))),
        (ps.gen, sk.gen, sl.gen, z)),
    ("quadrature",    (p,s,l,q) -> ebm.quad(ebm,p,s,l,q),
        (ps.ebm, sk.ebm, sl.ebm, sk.quad)),
    ("inv_transform", (m,p,s,l,r) -> m.sample_prior(m,p,s,l,r),
        (model, ps, sk, sl, sr)),
    ("log_prior",     (z,p,s,l,q) -> model.log_prior(z,ebm,p,s,l,q),
        (z_samp, ps.ebm, sk.ebm, sl.ebm, sk.quad)),
    ("generation",    model,
        (ps, sk, Lux.testmode(sl), sr)),
    ("train_step",    (o,p,s,l,x,i,r) -> model.train_step(o,p,s,l,x,i,r),
        (opt_st, ps, sk, sl, x, 1, sr)),
]

device = ENV["DEVICE"]
results = Dict()

for (name, f, args) in items
    _, t = bench(f, args...)

    hlo = string(Reactant.@code_hlo f(args...))
    profile = hlo_op_profile(hlo)

    results[name] = Dict(
        "timing_ms" => Dict("median" => median(t), "min" => minimum(t), "max" => maximum(t)),
        "hlo_ops" => [Dict("op" => op, "count" => n, "pct" => round(pct; digits=1)) for (op, n, pct) in profile],
        "total_ops" => sum(x[2] for x in profile),
    )

    @printf("%-20s  med=%.3f ms  min=%.3f ms  max=%.3f ms\n", name, median(t), minimum(t), maximum(t))
    print_hlo_profile(name, profile)
end

mkpath("profiles")
outfile = "profiles/profile_$(device).json"
open(outfile, "w") do io
    JSON.print(io, Dict("device" => device, "components" => results), 2)
end
println("\nSaved to $outfile")
```
